```python
data1 = pd.DataFrame({'key1' : [1,2,3,4,5,6],
                      'key2' : [4,5,6,7,8,9],
                      'key3' : [2,3,4,5,6,7]
})
```

```python
data2 = pd.DataFrame({'key1' : [1,2,3,45,67,6],
                      'key4' : [6,44,5,6,7,8],
                      'key5' : [23,4,5,8,6,7]
})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2)    #    Inner join
```

```python
pd.merge(data1,data2, how = 'left')    #    left join
```

```python
pd.merge(data1,data2, how = 'right')    #    right join
```

```python
pd.merge(data1,data2, how = 'outer')    #    outer join
```

```python
pd.merge(data1,data2, how = 'cross')    #    cross join (Maps everything with everything)
```

```python
pd.merge(data1,data2, how = 'outer',on = 'key1')
#    outer join
```

```python
data1 = pd.DataFrame({'key1' : [1,2,3,4,5,6],
                      'key2' : [4,5,6,7,8,9],
                      'key3' : [2,3,4,5,6,7]},
                      index = ['a','b','c','d','e','f']
)
```

```python
def cato_fare(x):
    if x<10:
        return "cheap"
    elif x>=10 and x<20:
        return 'mid'
    else:
        return 'high'
```

```python
df['cato_fare'] = df['Fare'].apply(cato_fare)
```
