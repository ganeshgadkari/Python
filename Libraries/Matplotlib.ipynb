```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
```

```python
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y , c = colours, s = sizes, alpha = .3)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")
```

```python
x = ['a','b','c','d','e']
```

```python
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Bar Graph")
```

```python
plt.barh(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Bar Graph")
```

```python
plt.figure(figsize = (6,2))
plt.plot(x,y,'--r')
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y,c = 'b', s = sizes, alpha = .5, marker = 'x')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter Plot")
```
